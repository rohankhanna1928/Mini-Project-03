# **1. Problem Statement:**
Find all the subsets from a set of numbers whose sum is zero.

Constraint: Subset size must be 5

Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
import random as r
Set = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
SetSize = 5
ResultList = set()
Iterations = 100000
for i in range(Iterations):
    S = r.sample(Set, SetSize)
    S.sort()
    if sum(S) == 0:
        ResultList.add(tuple(S))
for result in sorted(ResultList):
    print(result)
print("Total Unique Sets Found:", len(ResultList))
```

# **2. Problem Statement:**
Find all the subsets from a set of numbers whose sum is zero.

Constraint: Subset size must be 3 to 6 only

Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
import random as r
Set = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
SetLB = 3
SetUB = 6
ResultList = set()
Iterations = 1000000
for i in range(Iterations):
    SetSize = r.randint(SetLB,SetUB)
    S = r.sample(Set,SetSize)
    S.sort()
    if sum(S) == 0:
        ResultList.add(tuple(S))
for r in ResultList:
	print (r)
print ("Total Sets: ", len(ResultList))
```
